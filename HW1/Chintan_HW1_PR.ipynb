## Identity Matrix

An identity matrix is a given square matrix of any order which contains on its main diagonal elements with value of one, while the rest of the matrix elements are equal to zero.

The product of any square matrix and the appropriate identity matrix is always the original matrix, regardless of the order in which the multiplication was performed! 
In other words, AI=IA=A.

## Part 4

We are going to use `identity_matrix` to scramble the rows and columns of `img_average_channel`. 

1. First, create a numpy array that contains the numbers $0,1, \cdots, N$ where $N$ is the number of rows in the image.

2. Use `numpy.random.shuffle` to shuffle the indices. Call the shuffled array, `rows`

4. Print `rows` and inspect the elements.

```python
# Get the number of rows in img_average_channel
num_rows = img_average_channel.shape[0]

# Create an array of indices from 0 to num_rows-1
indices = np.arange(num_rows)

# Shuffle the indices using numpy.random.shuffle
np.random.shuffle(indices)

# The shuffled indices are now in the 'indices' array
rows = indices

# Print the shuffled rows array to inspect the elements
print("Shuffled Rows:")
print(rows)
```

When we print the "indices" array, we will see that it contains the numbers from 0 to N-1 in a shuffled order. These shuffled indices can be used to access or manipulate rows of data in a random order, which can be useful in various applications like data augmentation or random sampling.

## Part 5

1. Use your array of random row indices to create a new matrix `T` from `identity_matrix`. Overall, `T` has the same elements as `identity_matrix`, but the order of the rows of `T` should be according to the variable `rows`
2. Take a moment to think about what we are doing here. Describe in your own words what matrix `T` is.
3. Sanity check, print the shape of `T` and make sure it is the same size as `identity_matrix` and `img_average_channel`. If you need to,  reshape it to have the same size.


```python
# Create matrix T by rearranging rows from the identity_matrix
T = identity_matrix[rows]

# Sanity check: Print the shape of T and make sure it matches identity_matrix and img_average_channel
print("Shape of T:", T.shape)

# If needed, reshape T to match the shape of img_average_channel
if T.shape != img_average_channel.shape:
    T = T.reshape(img_average_channel.shape)

# Verify that T now has the same shape as identity_matrix and img_average_channel
print("Shape of T after reshaping:", T.shape)
```

Matrix T is essentially the identity matrix with its rows rearranged according to the random order specified by the "rows" array. Each row of T is still an identity vector with a single "1" at a specific position, but the order of these rows has been shuffled.

## Part 6

1. Matrix multiply `T` and `img_average_channel` and store the result in `transformed_image`. Check the shape of the result.
2. Plot `transformed_image` and describe the pattern you see. 
3. Think about what each dot product in the matrix multiplication is doing. Describe in your own words what this transformation is doing to the original image. 

```python
# Perform matrix multiplication between T and img_average_channel
transformed_image = np.dot(T, img_average_channel)

# Check the shape of the resulting transformed_image
print("Shape of Transformed Image:", transformed_image.shape)

# Plot the transformed_image
plt.imshow(transformed_image, cmap='gray')
plt.title("Transformed Image")
plt.axis('off')
plt.show()
```

In the transformed image, we may observe a specific pattern or structure. Each row of T corresponds to a specific row of the identity matrix, which represents a unit vector with a single "1" at a particular position.

When we perform matrix multiplication, each row of T effectively selects a specific row from the img_average_channel and scales it by 1. This means that each row in the transformed image is a row from img_average_channel that has been selected and retained in a specific order dictated by the shuffled indices.

So, the transformation rearranges the rows of the img_average_channel based on the shuffled order specified by the "rows" array. This results in a new image where the rows from the original image have been rearranged, creating a different pattern or structure compared to the original image. The specific pattern will depend on the shuffled order of rows, which is random.

## Part 7

1. We now want to retrieve the original image. Is matrix `T` invertible? How would you show that it is invertible?
2. Compute the inverse and store the matrix in `Tinv`.
3. Perform the inverse of the scrambling operation and plot the result to show that you retrieved the original image.

```python
# Step 1: Check if matrix T is invertible
det_T = np.linalg.det(T)

if det_T != 0:
    # Step 2: Compute the inverse matrix Tinv
    Tinv = np.linalg.inv(T)

    # Step 3: Apply the inverse operation to retrieve the original image
    original_image = np.dot(Tinv, transformed_image)

    # Step 4: Plot the original image
    plt.imshow(original_image, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')
    plt.show()
else:
    print("Matrix T is not invertible, so the original image cannot be retrieved.")
```

## Extra Points

* Now can you perform the scrambling on __both the columns and rows__ of the original image?  Write the code and plot the scrambled image. Then invert the operations and show that you can recover the original image. 

```python
# Scramble both rows and columns
num_rows, num_cols = img_average_channel.shape
row_indices = np.arange(num_rows)
col_indices = np.arange(num_cols)
np.random.shuffle(row_indices)
np.random.shuffle(col_indices)

# Scramble the original image
scrambled_image = img_average_channel[row_indices][:, col_indices]

# Plot the scrambled image
plt.imshow(scrambled_image, cmap='gray')
plt.title("Scrambled Image")
plt.axis('off')
plt.show()

# Invert the scrambling operations
# Create the inverse row and column indices
inverse_row_indices = np.argsort(row_indices)
inverse_col_indices = np.argsort(col_indices)

# Unscramble the scrambled image
recovered_image = scrambled_image[inverse_row_indices][:, inverse_col_indices]

# Plot the recovered original image
plt.imshow(recovered_image, cmap='gray')
plt.title("Recovered Original Image")
plt.axis('off')
plt.show()
```
